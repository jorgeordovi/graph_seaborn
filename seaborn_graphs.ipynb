```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North','South', 'East', 'West'],
    'Month': ['January', 'January', 'January', 'January', 'February', 'February', 'February', 'February'],
    'Sales': [100,150,120,110,130,140,160,155]
}
```

```python
df = pd.DataFrame(data)

sns.barplot(x='Region', y='Sales', data=df, hue='Month')
plt.title('Sales by Region & Month')
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(10,6))

sns.barplot(x='day', y='tip', data=tips, palette='Set2')

plt.title('Average tips by day of the week')
plt.xlabel('Days of the week')
plt.ylabel('Average Tip')

plt.show()
```
