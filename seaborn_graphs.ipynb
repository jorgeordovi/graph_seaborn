```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North','South', 'East', 'West'],
    'Month': ['January', 'January', 'January', 'January', 'February', 'February', 'February', 'February'],
    'Sales': [100,150,120,110,130,140,160,155]
}
```

```python
df = pd.DataFrame(data)

sns.barplot(x='Region', y='Sales', data=df, hue='Month')
plt.title('Sales by Region & Month')
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(10,6))

sns.barplot(x='day', y='tip', data=tips, palette='Set2')

plt.title('Average tips by day of the week')
plt.xlabel('Days of the week')
plt.ylabel('Average Tip')

plt.show()
```

```python
sns.get_dataset_names()
```

```python
sns.histplot(tips['total_bill'], kde=True, bins=30, color='skyblue')
plt.title('Total count with KDE distribution')
plt.xlabel('Total count in ($)')
plt.ylabel('Frequency')

plt.show()
```

```python
import scipy as sp
```

```python
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue')
plt.title('Estimated nucleus density KDE distribution')
plt.xlabel('Total count in ($)')
plt.ylabel('Frequency')
```

```python
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue', bw_adjust=0.5)
plt.title('Estimated nucleus density KDE distribution')
plt.xlabel('Total count in ($)')
plt.ylabel('Frequency')
```

```python
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue', cumulative=True)
plt.title('Estimated nucleus density KDE distribution')
plt.xlabel('Total count in ($)')
plt.ylabel('Frequency')
```

```python
sns.kdeplot(x=tips['total_bill'], y=tips['tip'], fill=True, cmap='Blues')
plt.title('Estimated Bi-variated of total tip distribution')
plt.xlabel('Total count in ($)')
plt.ylabel('Frequency')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', marginal_kws=dict(bins=15, fill=True))

plt.show()
```

```python
# Regression Model

sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')

plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')

plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')

plt.show()
```

```python
sns.swarmplot(
    data=tips.query("size in [2, 3, 5]"),
    x='total_bill',y='size', orient='h',
    native_scale=True,
)
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='day', y='tip', data=tips)

plt.title('Tip distribution per day')
plt.xlabel('Day')
plt.ylabel('Amount')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')

sns.swarmplot(x='day', y='tip', hue='sex', data=tips, palette='Set1', dodge=True)


plt.title('Tip disptributoion by day separated by sex')
plt.xlabel('Day')
plt.ylabel('Tip Amount')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[:2], labels[:2])

plt.show()
```
